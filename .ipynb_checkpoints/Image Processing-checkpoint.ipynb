```python
# loop over the colors dictionary
def subtract_lighter_pixels(color):
    b, g, r = color
    if b > 198 and g > 198 and r > 198:
        return 255, 255, 255
    return b, g, r

def validate_image_row(row_array):
    return [subtract_lighter_pixels(pixel) for pixel in row_array]
    
print(img_sample.shape)
editted_image = [validate_image_row(row) for row in img_sample]

# print(editted_image)
    
```

```python
from collections import OrderedDict
colors = OrderedDict({
    "red": (255, 0, 0),
    "green": (0, 255, 0),
    "blue": (0, 0, 255)})

lab = np.zeros((len(colors), 1, 3), dtype="uint8")

print(colors)
print(lab)

colorNames = []
# loop over the colors dictionary
for (i, (name, rgb)) in enumerate(colors.items()):
    # update the L*a*b* array and the color names list
    lab[i] = rgb
    colorNames.append(name)
    
print(colorNames)
print(lab)
```
