## Predicting the price of Avocado using NN

```python
import pandas as pd
import numpy as np
# Make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from sklearn.model_selection import train_test_split

# from keras.preprocessing.image import load_img, array_to_img
# from keras.utils.np_utils import to_categorical
# from keras.models import Sequential
# from keras.layers import Dense
# # Dense layer is used to achieve the fully connected neural networks

# import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import and Explore Avocado Dataset

```python
dataframe = pd.read_csv("avocado.csv")

dataframe.columns = [
    "index", "date", "price", "volume", "4046", "4225", "4770", 
    "total_bags", "small_bags", "large_bags", "xlarge_bags", "type", "year", "region"]
print(dataframe.shape)
dataframe.head()
```

```python
dataframe.describe()
```

```python
df_subset = dataframe.filter([
    "price", "volume", "4046", "4225", "4770", 
    "total_bags", "small_bags", "large_bags", "xlarge_bags", "type", "year", "region"], axis=1)

df_subset.dtypes
```

```python
# Check if we have any null fields
df_subset[df_subset.isnull().any(axis=1)]
```

```python
df_subset["region"].value_counts()
df_subset["type"].value_counts()
```

```python
df_encoded = pd.get_dummies(df_subset, columns=["region", "type"])
print(df_encoded.shape)
df_encoded.head()
```

```python
df_label = df_encoded.pop("price")
print(df_label.shape)
print(df_encoded.shape)
df_encoded.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_encoded, df_label, test_size=0.33)
```

```python
normalize = preprocessing.Normalization()
normalize.adapt(X_train)

avocado_model = tf.keras.Sequential([
  normalize,
  layers.Dense(64),
  layers.Dense(1)
])

avocado_model.compile(loss = tf.losses.MeanSquaredError(),
                      optimizer = tf.optimizers.Adam())

model_fit_history = avocado_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
# print(model_fit_history)
prediction = avocado_model.predict(X_test)
```

```python
def get_average_chunk(data):
    groups = [data[x:x+100] for x in range(0, len(data), 100)]

    chunked_mean = [sum(group)/len(group) for group in groups]
    return chunked_mean

plt.plot(get_average_chunk(y_test), color = 'blue')
plt.plot(get_average_chunk(prediction), color = 'red')
```
